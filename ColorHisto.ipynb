```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

Lecture de l'image. (Image provenant de PETS 2006, http://www.cvg.reading.ac.uk/PETS2006/data.html)

```python
image = cv2.imread('images/in000235.jpg') #OpenCV lit les images en BGR
b,g,r = cv2.split(image)
image = cv2.merge([r,g,b])
plt.imshow(image)
plt.show()
```

Calcul de l'histogramme de chaque canal.

```python
hist = cv2.calcHist([image],[0],None,[256],[0,256])
plt.plot(hist,'r')
hist = cv2.calcHist([image],[1],None,[256],[0,256])
plt.plot(hist,'g')
hist = cv2.calcHist([image],[2],None,[256],[0,256])
plt.plot(hist,'b')
plt.show()
```
