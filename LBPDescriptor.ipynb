Exemple d'utilisation de LBP

```python
import numpy as np
from skimage import data
from skimage.feature.texture import local_binary_pattern
import matplotlib.pyplot as plt
```

Obtention d'une image de skimage. Affichage.

```python
image = data.camera() 
plt.figure(figsize=(6,6)) 
plt.imshow(image, cmap='gray')
plt.show()
```

Application de LBP sur l'image pour des codes d'une taille donn√©e et pour un rayon pr√©d√©fini.

```python
longueurCodes = 8
rayon = 1
codes = local_binary_pattern(image, longueurCodes, rayon)
```

Analyse du r√©sultat sur une portion d'image.

```python
print(image[270:278,200:208])
```

Codes obtenus pour cette portion d'image.

```python
print(codes[270:278,200:208])
```

Les codes affich√©s sont les cha√Ænes $longueurCodes$ bits en entiers non-sign√©s. On peut d√©tailler les codes avec:

```python
code = 30
"{0:08b}".format(code)
```

Normalement, on fait des statistiques sur le nombre de r√©p√©titions des codes. Et cela nous donne le descripteur.

```python
hist, _ = np.histogram(codes, 256, range=(0, 256))
plt.plot(hist)
plt.show()
```
