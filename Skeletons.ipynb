Exemple du calcul d'un squelette

```python
from skimage.morphology import skeletonize
from skimage import data
import numpy as np
import matplotlib.pyplot as plt
```

Utilisation de l'image "horse" de skimage. Conversion en uint8 pour l'extraction du squelette. Affichage de l'image "horse".

```python
image = np.invert(data.horse()) #Le squelette est extrait sur la partie à True (noir sur l'image).
imageuint = image.astype('uint8') * 255 #Pour utiliser la fonction thinning de opencv.
plt.figure(figsize=(8,8))
plt.imshow(image, plt.get_cmap('binary'))
plt.show()
```

Calcul du squelette et affichage du résultat.

```python
squelette = skeletonize(imageuint)
plt.figure(figsize=(8,8))
plt.imshow(squelette, plt.get_cmap('binary'))
plt.show()
```

Ajout de "bruit" sur l'image originale.

```python
image[85:95,200:220] = True
image[125:130,120:140] = False
imageuint = image.astype('uint8') * 255 #Pour utiliser la fonction thinning de opencv.
plt.figure(figsize=(8,8))
plt.imshow(image, plt.get_cmap('binary'))
plt.show()
```

Résultat sur l'image bruité. Le squelette est très différent. Le bruit a une grande influence.

```python
squelette = skeletonize(imageuint)
plt.figure(figsize=(8,8))
plt.imshow(squelette, plt.get_cmap('binary'))
plt.show()
```
